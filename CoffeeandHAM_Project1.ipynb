### Items

```
#imports
import random


#start of our classes
#Make sure to run this section first or main wont compile

#-------------------------------------------------------------------------------
#Items

class Weapon:
    def __str__(self):
        return self.name

class Laptop:
    def __init__(self):
        self.name = "Laptop"
        self.description = "your laptop that does not work...but its pretty heavy"
        self.dmg = 10

    def __str__(self):
            return self.name
class Lint:
    def __init__(self):
        self.name = "Lint"
        self.description = "A piece of lint"
        self.dmg = 1

    def __str__(self):
        return self.name
class Toast(Weapon):
    def __init__(self):
        self.name = "Toast"
        self.description = "A piece of toast"
        self.dmg = 100

#-------------------------------------------------------------------------------
#Player

class Player:

    def __init__(self):
        self.backpack = [Toast()]
        self.x = 1
        self.y = 1 

    def print_backpack(self):
        print("Backpack: ")
        for item in self.backpack:
            print('* ' + str(item))
           
    
    def move(self, dirx, diry):
        self.x += dirx
        self.y += diry
        
    def move_north(self):
        self.move(dirx = 0, diry = -1)
    def move_south(self):
        self.move(dirx = 0, diry = 1)
    def move_west(self):
        self.move(dirx = -1, diry = 0)
    def move_east(self):
        self.move(dirx = 1, diry = 0)
#-------------------------------------------------------------------------------
#Enemies
class Enemy:
    def __init__(self):
        raise NotImplementedError("Do not create raw objects")
    
    def __str__(self):
        return self.name
    
    def am_i_alive(self):
        return self.hp > 0

class Dietrich(Enemy):
    def __init__(self):
        self.name = "Wild Dietrich"
        self.hp = 100
        self.damage = 50

class Erick(Enemy):
    def __init__(self):
        self.name = "Angry Erick"
        self.hp = 100
        self.damage = 50
#-------------------------------------------------------------------------------
#Our map
#actions correspond to Up (y - 1), Down (y + 1)
#Left (x - 1), and Right (x + 1)

#each tile holds a 'story' line
class FlatEarth:
  
    def __init__(self, x, y):
        self.x = x
        self.y = y
#will display the text from the tile you are on
    def text_from_tile(self):
        if self.enemy.am_i_alive():
            return "An {} appears".format(self.enemy.name)
        else:
            "u kill the monster"
        #raise NotImplementedError("Create a subclass instead!")

class StartTile(FlatEarth):
    def text_from_tile(self):
        return """
        you are in a dark an cold room
        """
class ComputerLab(FlatEarth):
    def text_from_tile(self):
        return """
        The computer lab... it is dark, you cant see. 
        
        """
class LoadUp(FlatEarth):
    def text_from_tile(self):
        return """
        On the table you see your laptop and a piece of toast
        
        """
class DeathTile(FlatEarth):
    def __init__(self, x, y):
        r = random.random()
        if r < 0.10:
          self.enemy = Dietrich()
        elif r < 0.50:
          self.enemy = Erick()
        else:
          self.enemy = Erick()
        
        super().__init__(x, y)




# |0,0||1,0||2,0|
# |0,1||1,1||2,1|
# |0,2||1,2||2,2|
# |0,3||1,3||2,3|

flatearth_map = [
    [StartTile(0,0),LoadUp(1,0),StartTile(2,0)],
    [DeathTile(0,1),StartTile(1,1),LoadUp(2,1)],             
    [StartTile(0,2),StartTile(1,2),StartTile(2,2)],
    [StartTile(0,3),ComputerLab(1,3),ComputerLab(2,3)],
]

def tile_location(x, y):
    if x < 0 or y < 0:
        return None
        
    try:
        return flatearth_map[y][x]
    except IndexError:
        return None
#-------------------------------------------------------------------------------
```

### Game Play

```


#function to move around game
def gameplay():

    print("The Last Vaquero Standing")
    
    #creating player obect from Player class
    player = Player()

    #backpack = [Laptop(),Lint(),Toast(), 'Laptop', 'Lint']
    
    #The folling code gets the users input and action is performed

    #this is our gameloop, without it our game and actions would only run once!
    #how to quit the game?
    
    while True: 
        roomIn = tile_location(player.x, player.y)
        
        print(roomIn.text_from_tile())

        player_input = get_player_input()
        
        #this controls the direction of the player
        if player_input == 'n' or player_input == 'N':
            player.move_north()
        elif player_input in ['s', 'S', 'south', 'South']:
            player.move_south()
        elif player_input in ['w', 'W', 'west', 'West']:
            player.move_west()
        elif player_input == 'e' or player_input == 'E':
            player.move_east()

        #other iputs for the game will go here

        #this displays items in your backpack aka inventory
        elif player_input in ['b', 'B', 'backpack', 'Backpack']:
            player.print_backpack()
            


        else:
            print("no entiendo")

#function to get players input
def get_player_input():
    return input('Action:')


gameplay()

```
